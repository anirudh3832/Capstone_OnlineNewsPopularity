## EDA Bivariate Analysis

```python
def groups(data,col):
    grp1=data.groupby(by=col).quantile(0.80)['shares']
    grp2=data.groupby(by=[col,'status']).quantile(0.75)['shares'].reset_index()
    popular=grp2[grp2['status']=='Popular']
    not_popular=grp2[grp2['status']=='Not Popular']
    return grp1,grp2,popular,not_popular


def line_plot(data,col):
    grp1,grp2,popular,not_popular=groups(data,col)
    sns.lineplot(x=grp1.index,y=grp1.values,markers='o')
    plt.title('Number of shares vs '+col)
    
def line_plot_hue(data,col):
    grp1,grp2,popular,not_popular=groups(data,col)
    sns.lineplot(x=grp2[col],y=grp2.shares,markers='o')
    sns.lineplot(x=popular[col],y=popular.shares,markers='o',label='popular')
    sns.lineplot(x=not_popular[col],y=not_popular.shares,markers='o',label='not popular')
    plt.title('Number of vs '+col+' vs Popularity')
    plt.legend(loc=1)
    plt.show()
    
```
