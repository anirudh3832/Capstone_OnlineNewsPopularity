## EDA Bivariate Analysis

```python
def groups(data,col):
    grp1=data.groupby(by=col).quantile(0.80)['shares']
    grp2=data.groupby(by=[col,'status']).quantile(0.75)['shares'].reset_index()
    popular=grp2[grp2['status']=='Popular']
    not_popular=grp2[grp2['status']=='Not Popular']
    return grp1,grp2,popular,not_popular


def line_plot(data,col):
    grp1,grp2,popular,not_popular=groups(data,col)
    sns.lineplot(x=grp1.index,y=grp1.values,markers='o')
    plt.title('Number of shares vs '+col)
    
def line_plot_hue(data,col):
    grp1,grp2,popular,not_popular=groups(data,col)
    sns.lineplot(x=grp2[col],y=grp2.shares,markers='o')
    sns.lineplot(x=popular[col],y=popular.shares,markers='o',label='popular')
    sns.lineplot(x=not_popular[col],y=not_popular.shares,markers='o',label='not popular')
    plt.title('Number of vs '+col+' vs Popularity')
    plt.legend(loc=1)
    plt.show()
    
```

```python
def splitting(data,col):
    X=data.drop(col,axis=1)
    y=(data[col])
    from sklearn.model_selection import train_test_split
    Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=1)
    return Xtrain,Xtest,ytrain,ytest
```

```python
def scaling(data,col):
    sc=StandardScaler()
    X=data.drop(col,axis=1)
    y=(data[col])
    from sklearn.model_selection import train_test_split
    Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=1)
    Xtrain=sc.fit_transform(Xtrain)
    Xtest=sc.transform(Xtest)
    return Xtrain,Xtest,ytrain,ytest
```

```python
def reg(data,col,func,flag=0):
    Xtrain,Xtest,ytrain,ytest=splitting(data,col) 
    if (flag==1) :
        Xtrain,Xtest,ytrain,ytest=scaling(data,col)        
    func.fit(Xtrain,ytrain)
    ypred_test=func.predict(Xtest)
    print('Testing Error')
    mse=mean_squared_error(ytest,ypred_test)
    rmse=np.sqrt(mse)
    print("MSE: ",mse)
    print("RMSE: ",rmse)
    rsquare=r2_score(ytest,ypred_test)
    print('R2score',rsquare)
    mae=mean_absolute_error(ytest,ypred_test)
    print("MAE: ",mae)
    mape=np.sum((np.abs((ypred_test-ytest))/ytest)*100)/len(ytest)
    print("MAPE: ",mape)
    print("")
    
    print('Training Error')
    ypredtrain=func.predict(Xtrain)
    mse=mean_squared_error(ytrain,ypredtrain)
    rmse=np.sqrt(mse)
    print("MSE: ",mse)
    print("RMSE: ",rmse)
    rsquare=r2_score(ytrain,ypredtrain)
    print('R2score',rsquare)
    mae=mean_absolute_error(ytrain,ypredtrain)
    print("MAE: ",mae)
    mape=np.sum((np.abs((ypredtrain-ytrain))/(ytrain+1))*100)/len(ytrain)
    print("MAPE: ",mape)
```
