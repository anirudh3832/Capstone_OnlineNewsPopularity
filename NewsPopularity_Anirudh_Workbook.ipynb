## Data Cleaning

It is found that column names are not defined properly 

example: ' timedelta'

we had an extra space in the name which made accessing data difficult

Therefore we have renamed it by using below method

```python
new_cols=[]
for col in data.columns:
    new_cols.append(col.strip())
```

```python
data.columns=new_cols
```

```python
data.columns
```

```python
int_col=['timedelta', 'n_tokens_title', 'n_tokens_content','num_hrefs', 'num_self_hrefs', 'num_imgs', 'num_videos','num_keywords', 'data_channel_is_lifestyle',
       'data_channel_is_entertainment', 'data_channel_is_bus',
       'data_channel_is_socmed', 'data_channel_is_tech',
       'data_channel_is_world','kw_min_min', 'kw_max_min', 'kw_avg_min',
       'kw_min_max', 'kw_max_max', 'kw_avg_max', 'kw_min_avg', 'kw_max_avg',
       'kw_avg_avg', 'self_reference_min_shares', 'self_reference_max_shares',
       'weekday_is_monday', 'weekday_is_tuesday',
       'weekday_is_wednesday', 'weekday_is_thursday', 'weekday_is_friday',
       'weekday_is_saturday', 'weekday_is_sunday', 'is_weekend','shares']
```

```python
for col in int_col:
    data[col]=data[col].astype('int32')
```

#### Number Of Links (num_hrefs)

```python
data['New_num_hrefs']=data['num_hrefs']
data['New_num_hrefs']=data['New_num_hrefs'].apply(lambda x : x if x<35 else 35)
```

```python
plt.figure(figsize=(20,5))
sns.countplot(data['New_num_hrefs'])
plt.title("Distribution of Number of Hyperlinks in the content")
plt.xlabel("No. of Hyperinks in Content")
plt.show()
```

#### Number of self referenced links (num_self_hrefs)

```python
data['New_num_self_hrefs']=data['num_self_hrefs']
data['New_num_self_hrefs']=data['New_num_self_hrefs'].apply(lambda x : x if x<25 else 25)
```

```python
plt.figure(figsize=(20,5))
sns.countplot(data['New_num_self_hrefs'])
plt.title("Distribution of Self Refrential Hyperlinks in the content")
plt.xlabel("No. of Self Refrential Hyperinks in Content")
plt.show()
```

#### Number of images (num_imgs)

```python
plt.figure(figsize=(20,5))
sns.countplot(data['New_num_imgs'])
plt.title("Distribution of Images in the content")
plt.xlabel("No. of Images in Content")
plt.show()
```

#### Number of videos (num_videos)

## Feature engineering

```python
data['New_Weekday']=data[['weekday_is_monday', 'weekday_is_tuesday',
       'weekday_is_wednesday', 'weekday_is_thursday', 'weekday_is_friday',
       'weekday_is_saturday', 'weekday_is_sunday', 'is_weekend']].idxmax(axis=1)
data['New_Weekday']=data['New_Weekday'].apply(lambda x : x[11:])
```

```python
data['New_Channel']=data[['data_channel_is_lifestyle',
       'data_channel_is_entertainment', 'data_channel_is_bus',
       'data_channel_is_socmed', 'data_channel_is_tech',
       'data_channel_is_world']].idxmax(axis=1)
data['New_Channel']=data['New_Channel'].apply(lambda x : x[16:])
```

```python
data.head()
```

```python
data['New_Channel'].value_counts()
```

```python
data['New_Channel']=data['New_Channel'].replace({'tech':'technology','bus':'business','socmed':'social media'})
```
